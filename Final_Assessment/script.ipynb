# [dbo].[GetCompanyDetailsById]

```sql
/****** Object:  StoredProcedure [dbo].[GetCompanyDetailsById]    Script Date: 13-12-2023 11:52:51 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[GetCompanyDetailsById]
  
AS
BEGIN
    SELECT
        Company_Table.*,
        Countries.CountryName,
        States.StateName,
        Cities.CityName
    FROM
        Company_Table
    INNER JOIN
        Countries ON Company_Table.Country = Countries.CountryId
    INNER JOIN
        States ON Company_Table.State = States.StateId
    INNER JOIN
        Cities ON Company_Table.City = Cities.CityId
   
END;
GO
```

# [dbo].[GetStatesByCountry]

```sql
/****** Object:  StoredProcedure [dbo].[GetStatesByCountry]    Script Date: 13-12-2023 11:52:51 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE PROCEDURE [dbo].[GetStatesByCountry]
    @CountryId INT = NULL
AS
BEGIN
    IF @CountryId =0
    BEGIN
        -- Return all states if CountryId is not provided
        SELECT StateId, StateName
        FROM States;
    END
    ELSE
    BEGIN
        -- Return states based on the specified CountryId
        SELECT StateId, StateName
        FROM States
        WHERE CountryId = @CountryId;
    END
END;
GO
```

# [dbo].[ValidateLoginUser]

```sql
/****** Object:  StoredProcedure [dbo].[ValidateLoginUser]    Script Date: 13-12-2023 11:52:51 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE PROCEDURE [dbo].[ValidateLoginUser]
    @Email NVARCHAR(255),
    @DecryptedPassword NVARCHAR(50),
    @Name NVARCHAR(255) OUTPUT
AS
BEGIN
    IF EXISTS (
        SELECT 1
        FROM UserTable
        WHERE Email = @Email
        AND decryptedPassword = @DecryptedPassword
        AND Name = @Name
    )
    BEGIN
        SELECT
            su.UserId,
            su.Name,
            su.Email
        FROM UserTable su
        WHERE su.Email = @Email
            AND su.decryptedPassword = @DecryptedPassword
            AND su.Name = @Name;

        -- Set the @name parameter value
        SET @Name = (SELECT su.Name FROM UserTable su WHERE su.Email = @Email AND su.decryptedPassword = @DecryptedPassword AND su.Name = @Name);
    END
END;
GO
```
